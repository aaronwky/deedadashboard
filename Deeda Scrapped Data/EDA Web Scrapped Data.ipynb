```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('campaigns_data.csv')
df.head()
```

```python
# Grouping by charity_name and identifying the best campaigns by donation_amount for each charity
best_campaigns = df.loc[df.groupby('charity_name')['donation_amount'].idxmax()]

# Select the relevant columns: charity_name, campaign_name, and donation_amount
best_campaigns = best_campaigns[['charity_name', 'campaign_name', 'donation_amount']]

# Display or save the result
print(best_campaigns)
```

```python
# Grouping by charity_name and identifying the best campaigns by donation_amount for each charity
best_campaigns = df.loc[df.groupby('charity_name')['donation_amount'].idxmax()]

# Sort by donation_amount to make the chart more readable
best_campaigns = best_campaigns[['charity_name', 'campaign_name', 'donation_amount']].sort_values(by='donation_amount', ascending=False)

# Set the plot size
plt.figure(figsize=(10, 8))

# Create a barplot for the best campaigns by donation amount
sns.barplot(y='charity_name', x='donation_amount', data=best_campaigns, palette='viridis')

# Add titles and labels
plt.title('Highest Donatation from Each Charitiy by Donation Amount', fontsize=16)
plt.xlabel('Donation Amount', fontsize=12)
plt.ylabel('Charity Name', fontsize=12)

# Rotate x-axis labels if necessary
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Group by charity_name and identify the best campaigns by donation_amount for each charity
best_campaigns = df.loc[df.groupby('charity_name')['donation_amount'].idxmax()]

# Sort by donation_amount to make the chart more readable and select the top 10 charities
top_10_best_campaigns = best_campaigns[['charity_name', 'campaign_name', 'donation_amount']].sort_values(by='donation_amount', ascending=False).head(10)

# Set the plot size
plt.figure(figsize=(10, 8))

# Create a barplot for the best campaigns by donation amount for top 10 charities
sns.barplot(y='charity_name', x='donation_amount', data=top_10_best_campaigns, palette='viridis')

# Add titles and labels
plt.title('Top 10 Charities by Highest Donation Amount', fontsize=16)
plt.xlabel('Donation Amount', fontsize=12)
plt.ylabel('Charity Name', fontsize=12)

# Rotate x-axis labels if necessary
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Filter out campaigns that did not raise any money
filtered_campaigns = df[df['donation_amount'] > 0]

# Group by charity_name and find the best campaign by donation amount for each charity
best_campaigns_filtered = filtered_campaigns.loc[filtered_campaigns.groupby('charity_name')['donation_amount'].idxmax()]

# Create a dataframe with relevant columns
best_campaigns_filtered_df = best_campaigns_filtered[['charity_name', 'campaign_name', 'donation_amount']]


# Display the dataframe
print("Highest Donatation from Each Charitiy (Excluding Zero Donations):")
print(best_campaigns_filtered_df)
```

```python
# Plot the filtered best campaigns
plt.figure(figsize=(12, 8))
sns.barplot(y='charity_name', x='donation_amount', data=best_campaigns_filtered_df, palette='Blues_r')
plt.title('Highest Donatation from Each Charitiy (Excluding Zero Donations)', fontsize=16)
plt.xlabel('Donation Amount (in $)', fontsize=12)
plt.ylabel('Charity Name', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Grouping by charity_name and identifying the worst campaigns by donation_amount for each charity
worst_campaigns = df.loc[df.groupby('charity_name')['donation_amount'].idxmin()]

# Select the relevant columns: charity_name, campaign_name, and donation_amount
worst_campaigns_df = worst_campaigns[['charity_name', 'campaign_name', 'donation_amount']]

# Sort the dataframe by donation_amount (ascending order) for clarity
worst_campaigns_df = worst_campaigns_df.sort_values(by='donation_amount', ascending=True)

# Display or save the result
print(worst_campaigns_df)
```

```python
# Grouping by charity_name and identifying the worst campaigns by donation_amount for each charity
worst_campaigns = df.loc[df.groupby('charity_name')['donation_amount'].idxmin()]

# Sort by donation_amount to make the chart more readable
worst_campaigns = worst_campaigns[['charity_name', 'campaign_name', 'donation_amount']].sort_values(by='donation_amount', ascending=True)

# Set the plot size
plt.figure(figsize=(10, 8))

# Create a barplot for the worst campaigns by donation amount
sns.barplot(y='charity_name', x='donation_amount', data=worst_campaigns, palette='magma')

# Add titles and labels
plt.title('Lowest Donatation from Each Charitiy by Donation Amount', fontsize=16)
plt.xlabel('Donation Amount', fontsize=12)
plt.ylabel('Charity Name', fontsize=12)

# Rotate x-axis labels if necessary
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Filter out campaigns that did not raise any money
no_donation_campaigns = df[df['donation_amount'] == 0]

# Get unique charities that have at least one campaign with zero donations
charities_no_donation = no_donation_campaigns[['charity_name', 'campaign_name', 'donation_amount']].drop_duplicates(subset='charity_name')

# Set the plot size
plt.figure(figsize=(10, 8))

# Create a barplot for the charities with campaigns that did not raise any money
sns.barplot(y='charity_name', x='donation_amount', data=charities_no_donation, palette='magma')

# Add titles and labels
plt.title('Charities with Campaigns that Raised No Money', fontsize=16)
plt.xlabel('Donation Amount', fontsize=12)
plt.ylabel('Charity Name', fontsize=12)

# Rotate x-axis labels if necessary
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Filtering campaigns that did not raise any money
no_donation_campaigns = df[df['donation_amount'] == 0]

# Creating a dataframe with relevant columns
no_donation_campaigns_df = no_donation_campaigns[['charity_name', 'campaign_name', 'donation_amount']]

# Display the dataframe
print("Campaigns That Did Not Raise Any Money:")
print(no_donation_campaigns_df)
```

```python

# Filter out campaigns that did not raise any money
no_donation_campaigns = df[df['donation_amount'] == 0]

# Count the number of zero-earning campaigns per charity
no_donation_counts = no_donation_campaigns.groupby('charity_name')['campaign_name'].count().reset_index()
no_donation_counts.columns = ['charity_name', 'zero_earning_campaigns']

# Bar chart showing the number of campaigns that did not raise money for each charity
plt.figure(figsize=(12, 8))
sns.barplot(y='charity_name', x='zero_earning_campaigns', data=no_donation_counts, palette='Reds_r')
plt.title('Number of Campaigns that Did Not Raise Money for Each Charity', fontsize=16)
plt.xlabel('Number of Zero-Earning Campaigns', fontsize=12)
plt.ylabel('Charity Name', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Filter campaigns that raised money
earned_campaigns = df[df['donation_amount'] > 0]

# Count the number of campaigns that raised money per charity
campaigns_per_charity = earned_campaigns.groupby('charity_name')['campaign_name'].count().reset_index()
campaigns_per_charity.columns = ['charity_name', 'campaigns_with_earnings']

# Bar chart for charities with the number of campaigns that raised money
plt.figure(figsize=(12, 8))
sns.barplot(y='charity_name', x='campaigns_with_earnings', data=campaigns_per_charity, palette='Greens_r')
plt.title('Number of Campaigns per Charity that Managed to Raise Money', fontsize=16)
plt.xlabel('Number of Campaigns', fontsize=12)
plt.ylabel('Charity Name', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Count the total number of zero-earning campaigns
total_zero_earning = len(no_donation_campaigns)

# Count the total number of all campaigns
total_campaigns = len(df)

# Create data for the pie chart
labels = ['Campaigns with Donations', 'Campaigns with No Donations']
sizes = [total_campaigns - total_zero_earning, total_zero_earning]
colors = ['#66b3ff', '#ff6666']

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Proportion of Campaigns with and without Donations', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Count the number of campaigns that raised money (donation_amount > 0) and didn't raise money (donation_amount == 0) per charity
campaigns_grouped = df.groupby('charity_name').apply(lambda x: pd.Series({
    'campaigns_with_earnings': (x['donation_amount'] > 0).sum(),
    'campaigns_without_earnings': (x['donation_amount'] == 0).sum()
})).reset_index()

# Plotting the stacked bar chart
campaigns_grouped.set_index('charity_name', inplace=True)
campaigns_grouped.plot(kind='bar', stacked=True, figsize=(12, 8), color=['green', 'red'])

# Adding labels and title
plt.title('Number of Campaigns per Charity: With and Without Earnings', fontsize=16)
plt.xlabel('Charity Name', fontsize=12)
plt.ylabel('Number of Campaigns', fontsize=12)
plt.xticks(rotation=45)
plt.legend(['Campaigns with Earnings', 'Campaigns without Earnings'])
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create a new column 'avg_amt_donated' by dividing donation_amount by donations_count
df['avg_amt_donated_per_person'] = ( df['donation_amount'] / df['donations_count'] ).round(2)

# Creating a new DataFrame with the relevant columns
df1 = df[['charity_name', 'campaign_name', 'donation_amount', 'donations_count', 'avg_amt_donated_per_person']]

# Display the new DataFrame
print(df1.head())
```

```python
# Barplot for average amount donated per charity
plt.figure(figsize=(12, 8))
sns.barplot(y='charity_name', x='avg_amt_donated_per_person', data=df1, palette='Blues_r')
plt.title('Average Donation Amount for Each Charity', fontsize=16)
plt.xlabel('Average Amount Donated (in $)', fontsize=12)
plt.ylabel('Charity Name', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Histogram: Distribution of comments_count
plt.figure(figsize=(8, 6))
sns.histplot(df['comments_count'], bins=20, kde=True, color='purple')
plt.title('Distribution of Comments Count', fontsize=16)
plt.xlabel('Comments Count', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#Scatter plot: Relationship between comments_count and donation_amount
plt.figure(figsize=(8, 6))
sns.scatterplot(x='comments_count', y='donation_amount', data=df, hue='charity_name', palette='Set1')
plt.title('Comments Count vs. Donation Amount', fontsize=16)
plt.xlabel('Comments Count', fontsize=12)
plt.ylabel('Donation Amount (in $)', fontsize=12)
plt.tight_layout()
plt.show()
```
