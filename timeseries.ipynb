```python
# save to json file 
import json
with open('aggregated data.json', 'r') as file:
    data = json.load(file)


donation_data_deeda = []
for charity in data:
    for campaign in charity['list_campaigns']:
        for donation in campaign['list_donations']:
            charity_item = {
            "charity_name" : charity['name'],
            "campaign_name" : campaign['title'],
            "donation_amount" : donation['amount']/100,
            'donateTime': donation['donateTime']
            }
            donation_data_deeda.append(charity_item)

      

```

```python
donation_data_deeda
```

```python
with open('donation_data_deeda.json', 'w') as outfile:
    json.dump(donation_data_deeda, outfile)
```
